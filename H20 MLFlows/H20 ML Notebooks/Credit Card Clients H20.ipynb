```python
import h2o
h2o.init()
```

```python
df_uci = h2o.import_file("D:\\DC Universe\\Ucsc\\Third Year\\ENH 3201 Industrial Placements\\H20 Applications\\Csv\\UCI_Credit_Card.csv")
```

```python
df_uci.columns
df_uci=df_uci.drop('ID')
```

```python
df_uci.types
```

```python
#train test valid split
train,test,valid = df_uci.split_frame(ratios=[.7, .15])
```

```python
y = "default.payment.next.month"
x = train.columns
x.remove(y)
train[y] = train[y].asfactor()
test[y] = test[y].asfactor()

```

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator
rf_fit1 = H2ORandomForestEstimator(model_id='rf_fit1', seed=1)
```

```python
rf_fit1.train(x=x, y=y, training_frame=train)
```

```python
#Model Performance
rf_performance = rf_fit1.model_performance(test)
print(rf_performance)
```

```python
#Test set AUC
print(rf_performance.auc)
```

```python
from h2o.automl import H2OAutoML
aml = H2OAutoML(max_models=5, max_runtime_secs=300, seed = 1)
aml.train(x = x, y = y, training_frame = train)
```

```python
lb = aml.leaderboard
lb.head()
```

```python
model = aml.leader
```
